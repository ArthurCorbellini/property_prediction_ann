```python
import math

# Abra o arquivo de entrada para leitura
with open('log_002-FINAL.txt', 'r') as arquivo_entrada:
    # Leia todas as linhas do arquivo de entrada
    rows = arquivo_entrada.readlines()

# Abra o arquivo de sada em formato CSV
data = []
obj = {}

is_current = False
for row in rows:
    if row.startswith('looping'):
        obj['looping'] = int(row.strip().replace('looping:  ', ''))
    if row.startswith('temp'):
        obj['temp'] = float(row.strip().replace('temp: ', ''))
    if row.startswith('acc_probability'):
        obj['acc_probability'] = float(row.strip().replace('acc_probability: ', ''))
    if row.startswith('new_become_current'):
        obj['new_become_current'] = row.strip().replace('new_become_current: ', '') == 'True'
    if row.startswith('training_time'):
        obj['training_time'] = row.strip().replace('training_time: ', '')

    if row.startswith('current'):
        is_current = True
    if is_current and row.startswith(' - mae: '):
        obj['current_mae'] = float(row.strip().replace('- mae: ', ''))
        is_current = False

    if row.startswith('-\\'):
        data.append(obj)
        print(obj)
        obj = {}
```

```python
import matplotlib.pyplot as plt

X = [item['looping'] for item in data]
Y = [item['temp'] for item in data]
plt.figure(figsize=(6, 3))

plt.plot(X, Y, linestyle='-')
# plt.title('Grfico de Temperatura x Looping')
plt.xlabel('looping')
plt.ylabel('temperature')
plt.grid(True)

plt.show()
```
